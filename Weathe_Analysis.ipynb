## Find the Weathers number whose temperature is greater than average temperature

```python
print("Average temperature from dataset ->",round(np.mean(data['Temp_C']),2))
print("Total number of data greater then average temperature-> ",data[(data['Temp_C']) > np.mean(data['Temp_C'])]['Temp_C'].count())
print("\n----------------------Total number of weathers occures which is greater then average temperature------------------")
weather_average_temp = data[(data['Temp_C']) > np.mean(data['Temp_C'])][['Temp_C','Weather']]
print(weather_average_temp.groupby('Weather').count().sort_values(by= 'Temp_C',ascending= False))
```

# Plot Visualisation

## Temperature vs Date/Time using plot chart

```python
plt.figure(figsize = (20,6))
plt.plot(data["Date/Time"],data["Temp_C"], color = "b",label="Temperature (Â°C)")
plt.title("Temaperature over time")
plt.xlabel("Date and Time")
plt.ylabel("Temperature")
plt.grid(True)
plt.xticks([])
plt.show()
```

## Correlation between every columns

```python
corr_matrix = data.corr()
corr_matrix
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',fmt=".2f",linewidths=0.5)
plt.title("Correlation data of weather dataset")
plt.tight_layout()
plt.plot()
```

## Bar Plot for top 10 occur Weather Conditions

```python
weather_count = data.groupby("Weather").count().sort_values(by="Temp_C",ascending=False)
top_10_weather_count = weather_count['Temp_C'][0:10]
plt.figure(figsize = (20,6))
ax = sns.barplot(x = top_10_weather_count.keys(), y= top_10_weather_count.values,palette='Set2')
ax.bar_label(ax.containers[0])
plt.show()
```

## Wind Speed vs Visibility Plot

```python
plt.figure(figsize = (15,8))
plt.scatter(x = data["Wind Speed_km/h"], y=data["Visibility_km"],color='tab:orange')
plt.title('Wind Speed vs Visibility')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Visibility (km)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Temperature vs Dew Point Scatter Plot

```python
plt.figure(figsize = (15,8))
sns.scatterplot(x = data["Temp_C"], y = data["Dew Point Temp_C"],palette="Set3")
plt.title('Temperature vs Dew Point Temperature')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Dew Point Temperature (Â°C)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
